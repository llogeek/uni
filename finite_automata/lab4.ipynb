```python
def find_in_classes(classes, cl):
    for i in range(len(classes)):
        if cl in classes[i]:
            return i

def build_classes(automata_psi):
    c = set(automata_psi.values())
    i = -1
    current_classes = [[] for _ in range(len(c))]
    for val in c:
        i = i + 1
        for key in automata_psi.keys():
            if automata_psi[key] == val:
                current_classes[i] += [key]
    return current_classes

def process_class(prev_classes, cls, automata_phi):
    class_states = {}
    for state in cls:
        seq = ''.join([str(find_in_classes(prev_classes, i)) for i in automata_phi[state]])
        class_states.update({state:seq})
    current_classes = build_classes(class_states)
    return current_classes

def build_temp_state(prev_classes, automata_phi, automata_psi):
    current_classes = []
    for clas in prev_classes:
        update_class = process_class(prev_classes, clas, automata_phi)    
        for i in update_class:
            current_classes.append(i)
    return current_classes
```
